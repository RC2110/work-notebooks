Deriving weighted average for each movies using the below expression.
We penalize the movies using the ones that have less vote count & average rating of movies, to derive the popularity scores for movies.

weighted_average = (vote_count/(vote_count + m))*R + (m /(vote_count+m))*C

m = number of movies that have less vote_count

C = average rating of all the movies

```python
def weighted_average(df, m=m, c=C):
    v=df['vote_count']
    R=df['vote_average']
    return (v/(v+m) *R) + (m /(v+m)*C)
```

```python
movies_filtered['weighted_average1'] = movies_filtered.apply(weighted_average, axis=1)
```
